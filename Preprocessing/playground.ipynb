```python
days_count = np.zeros(366 + 365)
for day in days:
    days_count[day] += 1
days_count = days_count / days_count.max()
plt.scatter(np.linspace(0, 366 + 365 - 1, 366 + 365), days_count, alpha=0.8, s=4 * (1.1 - days_count),
            c=["red" if d == 0 else "green" for d in days_count])
plt.xlabel("day")
plt.ylabel("appearance")
plt.title("Missing days are marked as 0")
plt.show()


def get_date(d):
    months_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    year = 2016 + d // 366
    if year == 2016:
        months_days[1] += 1
    d = d % 366
    for i in range(12):
        if d - months_days[i] > 0:
            d -= months_days[i]
        else:
            break
    month = i + 1
    day = d
    return str(year) + ":" + str(month).rjust(2) + ":" + str(day).rjust(2)


print("missing days are :")
missing_days = np.where(days_count == 0)[0] + 1
missing_days_date = []
for day in missing_days:
    missing_days_date.append(get_date(day))
print("\t-\t".join(missing_days_date))
```

# Sort columns

```python
def sort_columns(df, cols_to_set_first=("Id", "prediction", "current hour precipitations (kg/m^2)",
                                        "next hour precipitation (kg/m^2)",
                                        "station_id", "altitude (m)", "latitude", "longitude", "latitude_idx",
                                        "longitude_idx", "month", "month_cos", "month_sin", "day", "hour", "hour_cos",
                                        "hour_sin", "wind_speed (m/s)", "wind_direction_cos", "wind_direction_sin",
                                        "wind_direction (deg)", "wind_component_u (m/s)", "wind_component_v (m/s)")):
    cols = list(df.columns)
    cols_to_remove = []
    for c in list(cols_to_set_first):
        if c in cols:
            cols.remove(c)
        else:
            print(c, "not fount.")
            cols_to_remove.append(c)
    cols = list(cols_to_set_first) + cols
    for c in cols_to_remove:
        cols.remove(c)
    return cols


def save_file(x, save_path, index=False):
    # save in file
    if save_path:
        if not os.path.exists("/".join(save_path.split("/")[:-1])):
            os.mkdir("/".join(save_path.split("/")[:-1]))
        x.to_csv(save_path, index=index)
        print(f"file saved in '{save_path}'")
```
