```python
import umap
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import itertools
import numpy as np
from sklearn.model_selection import ParameterGrid
```

```python
def get_data(remove_irrelevant=True):
    data = pd.read_csv('/content/drive/MyDrive/weather_data/X_station_coord_2nn_imputed.csv')
    if remove_irrelevant:
        del data['number_sta']
        del data['Id']
        del data['hour']
        del data['latitude']
        del data['longitude']
        del data['height_sta']
        del data['ground_truth']
    del data['date']
    del data['timestamp']
    print(data.columns.tolist())
    return data
```

```python
parameters = ['wind_direction', 'humidity', 'dew_point', 'temperature', 'wind_speed', 'precip', 'month']
data = get_data(remove_irrelevant=True).sample(n=100000)

for i in range(len(parameters)):
    possible_combinations = itertools.combinations(parameters, i)
    reducer = umap.UMAP(random_state=42)
    for combination in possible_combinations:
        print(np.array(combination))
        if len(np.array(combination)) > 0 :
          scaled_data = StandardScaler().fit_transform(data[np.array(combination)])
          embedding = reducer.fit_transform(scaled_data)

          fig = plt.figure()
          plt.scatter(embedding[:, 0], embedding[:, 1])
          plt.title(combination)
          file_name = str(combination) + '.jpg'
          fig.savefig(file_name, dpi=150)


    scaled_data = StandardScaler().fit_transform(data[np.array(parameters)])
    embedding = reducer.fit_transform(scaled_data)
    fig = plt.figure()
    plt.scatter(embedding[:, 0], embedding[:, 1])
    plt.title(parameters, fontsize=24)
    file_name = str(parameters) + '.jpg'
    fig.savefig(file_name, dpi=150)
```

```python
if __name__ == '__main__':
    parameters = ['wind_direction', 'humidity', 'dew_point', 'temperature', 'wind_speed', 'precip'] # , 'month']
    data = get_data(remove_irrelevant=True).sample(n=50000)[np.array(parameters)]
    scaled_data = StandardScaler().fit_transform(data)

    possible_params_umap = {'n_neighbors': [2, 5, 15],
                            'spread': [0.5, 1.0, 2.0],
                            'min_dist': [0.01, 0.1, 0.5],
                            'metric': ['euclidean', 'manhattan', 'jaccard']}

    params_umap = list(ParameterGrid(possible_params_umap))

    for param_umap in params_umap:
        print(param_umap)
        reducer = umap.UMAP(n_neighbors=param_umap['n_neighbors'], spread=param_umap['spread'],
                            min_dist=param_umap['min_dist'], metric=param_umap['metric'],
                            random_state=42)
        embedding = reducer.fit_transform(scaled_data)

        fig = plt.figure()
        plt.scatter(embedding[:, 0], embedding[:, 1])
        title = 'umap_result/' + 'n_neighbors ' + str(param_umap['n_neighbors']) + ' spread ' + \
                str(param_umap['spread']) + ' min_dist ' + str(param_umap['min_dist']) + ' metric ' + \
                str(param_umap['metric'])

        plt.title(title)
        file_name = str(title) + '.jpg'
        fig.savefig(file_name, dpi=150)
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score

from datetime import datetime


def cv_kmeans():
    features = [ 'wind_speed', 'precip'] #['wind_direction', 'humidity', 'dew_point', 'temperature', 'wind_speed', 'precip']
    data = pd.read_csv('/content/drive/MyDrive/weather_data/X_station_coord_2nn_imputed_by_day.csv')[np.array(features)]

    dict_results = dict()
    for feature in features:
      data_feat = data[[feature]]
      if feature == 'wind_speed' : low = 7 else : low = 2
      for i in range(low, 13):
          print(feature, i)
          dict_results[feature] = {}
          dict_results[feature][i] = {}

          kmeans = KMeans(n_clusters=i, random_state=0).fit(data_feat)

          print("Kmeans computed", datetime.now().strftime("%H:%M:%S"))

          # Get the cluster labels to compute scores

          cluster_labels = kmeans.predict(data_feat)

          print("cluster_labels computed", datetime.now().strftime("%H:%M:%S"))


          silhouette = silhouette_score(data_feat, cluster_labels)
          dict_results[feature][i]['silhouettes'] = silhouette
          ch_score = calinski_harabasz_score(data_feat, cluster_labels)
          dict_results[feature][i]['c_h'] = ch_score

          print("scores computed", datetime.now().strftime("%H:%M:%S"))

          print(dict_results[feature][i])

          # Plot the clusters
          # transformed_data = kmeans.transform(data)
          # print(transformed_data)

    return dict_results

print(cv_kmeans())
```

### Silhouette score 

The notion of silhouette is introducedin [Rousseeuw, 1987] to measure the discrepancy between clusters. In our context, we propose to useit to evaluate the difference between our different sources and between the sources and the target.Each year studied corresponds to a different distribution then we consider each one as a cluster forthis experiment and we attribute a label to all samples of the year without regard to their labels.For two clusters A and B and a given sampleiof A, the silhouette is defined ass(i) =b(i)−a(i)max (a(i),b(i))witha(i) the average dissimilarity ofito all the other examples of A,b(i) = minB6=Ad(i,B),d(i,B)the average dissimilarity of i to all the other examples of B. Note that,−1≤s(i)≤1.  Ifs(i) = 1,example i belongs to the correct cluster.  On the contrary, ifs(i) =−1, sample i has been consideredas belonging to the wrong cluster:  A instead of B. Lastly, ifs(i) = 0, sample i could equally wellbelong to group A or B.

### Calinski and Harabasz score (CH-score)

We also consider the Calinski and Harabasz score (CH-score) [Calinski and J, 1974] which isa  measure  of  the  dispersion  between  and  inter-cluster.   The  higher  the  score  is,  the  more  welldefined the clusters are:  the samples within a cluster will be close to each other and far from theones of other clusters.  Nevertheless, this score is not bounded therefore, the CH-score can only beassessed in comparison with other CH-score.
